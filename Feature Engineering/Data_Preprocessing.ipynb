```python
def generate_responses(prompts, source, temp, num_responses=200):
    responses = []
    num_prompts = len(prompts)
    num_temps = len(temperatures)

    # Determine how many responses per combination
    responses_per_combination = 5
    openai.api_key = OPENAI_API_KEY
    for prompt in prompts:
      responses_prompt = []
      print(prompt)
      for _ in range(responses_per_combination):
        print("Generating response...")
        response = openai.chat.completions.create(
            model="gpt-4o-mini",
            messages=[{"role": "user", "content": f"{prompt} \n Use the Source test: {source}"}],
            max_tokens=150,
            temperature=temp,
            top_p=1.0,  
            frequency_penalty=0.2,
            presence_penalty=0.3
        )
        # print(response)
        content = response.choices[0].message.content.strip()

        # print(content)
        responses_prompt.append(content)
      save_responses_to_csv(responses_prompt, temp)
      print("Responses generated and saved to 'generated_responses.csv'")
    return responses
```

```python
import csv

def save_responses_to_csv(responses, temp, filename="generated_responses.csv"):
    
    header = [ "text", "generated"]

    
    with open(filename, mode='a', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        
        writer.writerow(header)
       
        for i, response in enumerate(responses, 1):
            writer.writerow([response, 1, "gpt-4o-mini", temp ])
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
# from numba import cuda

# @cuda.jit
def generate_responses_llama(prompts, source, temp, num_responses=100):
   
    model_name="meta-llama/Llama-3.2-3B"
    hf_token = "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"
    tokenizer = AutoTokenizer.from_pretrained(model_name, use_auth_token=hf_token)

    
    model = AutoModelForCausalLM.from_pretrained(model_name,
                                                 use_auth_token=hf_token,
                                                 device_map="auto")
    device = 0 if torch.cuda.is_available() else -1
    # model.to(device)
    
    pipe = pipeline(
        "text-generation",
        model=model,
        tokenizer=tokenizer,
        torch_dtype="auto",
        device_map="auto"
        )


    responses_per_combination = 1

    for prompt in prompts:
      responses = []
      for _ in range(responses_per_combination):
        formatted_prompt = (
                    f"{prompt}\n\nBased on the following source text, respond succinctly and avoid repetition:\n\n{source}"
                )
        output = pipe(
          formatted_prompt,
          do_sample=True,
          top_k=50,
          temperature=temp,
          max_new_tokens=400
        )

        print(output)
        content = output[0]['generated_text'].strip()

        if content.startswith(formatted_prompt):
            content = content[len(formatted_prompt):].strip()

        
        if len(content.split('\n')) > 2: 
            content = content.split('\n')[0].strip()

        responses.append(content)
      save_responses_to_csv(responses, temp, filename="llama-generated.csv")
      print("Responses generated and saved to 'llama-generated.csv'")
    return responses
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

generated_responses = pd.read_csv("data/generated_responses.csv")

texts = generated_responses["text"].tolist()

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(texts)

cosine_sim_matrix = cosine_similarity(tfidf_matrix)

to_remove = set()
for i in range(len(texts)):
    for j in range(i + 1, len(texts)):
        if cosine_sim_matrix[i, j] > 0.9: 
            to_remove.add(j)

filtered_responses = generated_responses.drop(list(to_remove)).reset_index(drop=True)

filtered_responses.to_csv("filtered_generated_responses.csv", index=False)

print(f"Filtered dataset saved. Original rows: {len(generated_responses)}, Filtered rows: {len(filtered_responses)}")
```
